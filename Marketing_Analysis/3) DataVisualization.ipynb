# Data Visualization

In this notebook we are visualizing the data with the help of matplotlib and seaborn in python

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#Reading file with unknown values dropped
cust_data=pd.read_csv('dropped.csv')
```

### Percentage Turnout Analysis 

```python
#Plotting piechart for customer turnout analysis
plt.figure(figsize=(6, 6))

# Custom colors for 'No' and 'Yes'
colors = ['#0147AB', '#19ff99']  

cust_data['Customer_Response'].value_counts().plot.pie(autopct='%1.1f%%', labels=['No', 'Yes'], colors=colors)
plt.title('Percentage Turnout of Marketing Campaign')
```

- Compared to industry standards, the percentage turnout of 12.7% is excellent.
- The factors contributing and percentage of contributions are to be further analyzed.

### Contact Mode Impact Analysis 

```python
#Count plot based on Contact Mode
plt.figure(figsize=(10,5))
ax=sns.countplot(x='contact', data=cust_data, hue='Customer_Response')
ax.set_xticklabels(ax.get_xticklabels(), rotation=0)
ax.set_title('Response Based on Contact Mode')
```

- It can be observed that cellular mode of contact has a higher conversion rate and higher number of users.
- This can be attributed to the rising trend of cellphone users as compared to telephone.

## Analysis of Attempts to Convert Customer into Depositor

## Campaign Analysis

```python
#Count plot based on Calls during Campaign 
plt.figure(figsize=(11,4))
ax=sns.countplot(x='campaign', data=cust_data, hue='Customer_Response')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
ax.set_title('Response Based on Campaign')
```

### Analysis Based on Month of Contact

```python
#Count plot based on Month of Contact
plt.figure(figsize=(10,5))
ax=sns.countplot(x='month', data=cust_data, hue='Customer_Response')
ax.set_xticklabels(ax.get_xticklabels(), rotation=0)
ax.set_title('Response Based on Month')
```

### 3 Month Average Interest Rates of Eurozone Banks Based Analysis

```python
#Count plot based on euribor3m
plt.figure(figsize = (10, 30))
sns.countplot(data=cust_data, x='euribor3m', hue='Customer_Response')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.ylabel('Count')
plt.title('Customer Response divided by 3 Month Average Interest Rates of Eurozone Banks')
plt.legend(title='Customer Response', loc='upper right')
plt.show()
```

- There are a lot of values for EURIBOR3m so the graph is too congested to show any trends.
- Only inference is that there are a higher number of prospective consumers at a higher EURIBOR3m rate.
- A better observation may be obtained upon bucketing the values.

```python
#Getting a summary of euribor3m data
summary = cust_data['euribor3m'].describe()
print(summary)
#Values to help bucket euribor3m values
```

```python
#Creating buckets
q1 = summary['25%']
q3 = summary['75%']
iqr = q3 - q1
lower_cut = q1 - 1.5 * iqr
upper_cut = q3 + 1.5 * iqr
bins = [lower_cut, q1, q3, upper_cut]
print(bins)
```

```python
#Loading column onto dataset
labels = ['Low', 'Medium','High']
cust_data['euribor3m_category'] = pd.cut(cust_data['euribor3m'], bins=bins, labels=labels)
```

```python
#Count plot based on bucketed euribor3m
plt.figure(figsize = (10, 30))
sns.countplot(data=cust_data, x='euribor3m_category', hue='Customer_Response')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.ylabel('Count')
plt.title('Customer Response divided by 3 Month Average Interest Rates of Eurozone Banks')
plt.legend(title='Customer Response', loc='upper right')
plt.show()
```

- It can be observed that there is a higher number of term-deposit subscribers at a low EURIBOR and it is reducing with increasing EURIBOR.
- When interest rates are low across multiple investment opportunities, it is possible that term investment schemes have the highest ROI.
- Term deposit schemes are risk free and provide monetary returns that are certain for a fixed period of time. This is a possible positive for people during slow economic periods.
